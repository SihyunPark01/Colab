```python
import pandas as pd
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.utils import to_categorical
from keras.preprocessing.sequence import pad_sequences
from konlpy.tag import Okt, Kkma
from wordcloud import WordCloud
import seaborn as sns
```

```python
#########################################################
# 형태소 분석
```

```python
# 꼬꼬마 형태소 분석기 생성
kkma = Kkma()

text = '나는 보리밥을 먹었다.'
print('명사 :',kkma.nouns(text))
print('형태소 :',kkma.morphs(text))
print('품사 :',kkma.pos(text))
```

```python
# Okt(Open Korean Text) 텍스트 분석기 생성
okt = Okt()

text = '나는 보리밥을 먹었다.'
print('명사 :',kkma.nouns(text))
print('형태소 :',kkma.morphs(text))
print('품사 :',kkma.pos(text))
```

```python
#############################################################
# 토큰화 (BoW)

text = ['You are so beautiful',
        'You are the best',
        'I love dog',
        'I drink coffee every morning',
        'I hate you']

tokenizer = Tokenizer()
tokenizer.fit_on_texts(text)
word_dic = tokenizer.word_index
word_dic
```

```python
# 텍스트 벡터화
sequences = tokenizer.texts_to_sequences(text)
sequences #텍스트를 수치화 시킨 것임
```

```python
# 패딩처리 (단어의 문장의 길이가 다 다르므로 학습에 문제 발생 가능성 높음 -> 패딩처리 필요)
padd_sequences = pad_sequences (sequences, padding='post') # padding값 0을 post(뒷쪽)에 채워넣어 모두 5글자로 길이를 맞춤.
padd_sequences
```
